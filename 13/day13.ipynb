## Part 1

Say the target is a vector $(x_0,y_0)^T$ and the $A$ button adds $(x_A,y_A)^T$ to the current value whereas the $B$ button adds $(x_B,y_B)^T$. Then, we wanty to find a solution $(a,b)$ to 
\begin{equation}
    a
    \begin{pmatrix}
    x_A\\y_A
    \end{pmatrix}
    + b
    \begin{pmatrix}
    x_B\\y_B
    \end{pmatrix}=
    \begin{pmatrix}
    x_0\\y_0
    \end{pmatrix}\Leftrightarrow
     \begin{pmatrix}
    x_A & x_B \\ y_A & y_B
    \end{pmatrix}
    \begin{pmatrix}
    a\\b
    \end{pmatrix}=
    \begin{pmatrix}
    x_0\\y_0
    \end{pmatrix}.
\end{equation}
If the matrix is invertible, the solution is simply
\begin{equation}
    \begin{pmatrix}
    a\\b
    \end{pmatrix}=
    \begin{pmatrix}
    x_A & x_B \\ y_A & y_B
    \end{pmatrix}^{-1}
    \begin{pmatrix}
    x_0\\y_0
    \end{pmatrix}.
\end{equation}
