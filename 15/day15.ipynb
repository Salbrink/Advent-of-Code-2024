## Part 1

```python

file1 = open('puzzle_input_15_map.txt')
curr_r, curr_c = -1, -1
row = 0
map =[]
for line in file1:
    submap = []
    col = 0
    for c in line:
        if c == '@': 
            curr_r = row
            curr_c = col
        submap.append(c)
        col += 1
    row += 1
    map.append(submap)

file2 = open('puzzle_input_15_instructions.txt')
instructions = []
for line in file2:
    for c in line: instructions.append(c)

while instructions:
    i = instructions.pop(0)
    dr = 1
    match i:
        case '^':
            if not map[curr_r-dr][curr_c] == '#':
                if map[curr_r-dr][curr_c] == '.':
                    map[curr_r][curr_c] = '.'
                    map[curr_r-dr][curr_c] = '@'
                    curr_r -= 1

                elif map[curr_r-dr][curr_c] == 'O':
                    while -1 < curr_r - dr and map[curr_r-dr][curr_c] == 'O': dr += 1
                    if map[curr_r-dr][curr_c] == '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r-1][curr_c] = '@'
                        for j in range(2,dr+1):
                            map[curr_r-j][curr_c] = 'O'
                        curr_r -= 1
        case '>':
            if not map[curr_r][curr_c+dr] == '#':
                if map[curr_r][curr_c+dr] == '.':
                    map[curr_r][curr_c] = '.'
                    map[curr_r][curr_c+dr] = '@'
                    curr_c += 1

                elif map[curr_r][curr_c+dr] == 'O':
                    while curr_c + dr < len(map[curr_r]) and map[curr_r][curr_c+dr] == 'O': dr += 1
                    if map[curr_r][curr_c+dr] == '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r][curr_c+1] = '@'
                        for j in range(2,dr+1):
                            map[curr_r][curr_c+j] = 'O'
                        curr_c += 1
        case 'v':
            if not map[curr_r+dr][curr_c] == '#':
                if map[curr_r+dr][curr_c] == '.':
                    map[curr_r][curr_c] = '.'
                    map[curr_r+dr][curr_c] = '@'
                    curr_r += 1

                elif map[curr_r+dr][curr_c] == 'O':
                    while curr_r + dr < len(map) and map[curr_r+dr][curr_c] == 'O': dr += 1
                    if map[curr_r+dr][curr_c] == '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r+1][curr_c] = '@'
                        for j in range(2,dr+1):
                            map[curr_r+j][curr_c] = 'O'
                        curr_r += 1
        case '<':
            if not map[curr_r][curr_c-dr] == '#':
                if map[curr_r][curr_c-dr] == '.':
                    map[curr_r][curr_c] = '.'
                    map[curr_r][curr_c-dr] = '@'
                    curr_c -= 1

                elif map[curr_r][curr_c-dr] == 'O':
                    while -1 < curr_c - dr and map[curr_r][curr_c-dr] == 'O': dr += 1
                    if map[curr_r][curr_c-dr] == '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r][curr_c-1] = '@'
                        for j in range(2,dr+1):
                            map[curr_r][curr_c-j] = 'O'
                        curr_c -= 1

GPS = 0
for r, submap in enumerate(map):
    for c, element in enumerate(submap):
        if element == 'O':
            GPS += 100*r + c
print('GPS coordinate: ', GPS)
```

## Part 2

```python

file1 = open('puzzle_input_15_map.txt')
curr_r, curr_c = -1, -1
row = 0
map =[]
for line in file1:
    submap = []
    col = 0
    for c in line:
        match c:
            case '#':
                submap.append('#')
                submap.append('#')
            case 'O': 
                submap.append('[')
                submap.append(']')
            case '.':
                submap.append('.')
                submap.append('.')
            case '@':
                submap.append('@')
                submap.append('.')
                curr_r = row
                curr_c = col
        col += 2
    row += 1
    map.append(submap)

file2 = open('puzzle_input_15_instructions.txt')
instructions = []
for line in file2:
    for c in line: 
        if c != '\n':
            instructions.append(c)

def check_if_box_free_to_move(boxX,boxY,map,up=-1):
    if map[boxX+up][boxY] == '#' or map[boxX+up][boxY+1] == '#': return False

    if map[boxX+up][boxY] == '.' and map[boxX+up][boxY+1] == '.': return True

    if map[boxX+up][boxY] == '[':
        return check_if_box_free_to_move(boxX+up,boxY,map,up)
    
    if map[boxX+up][boxY] == ']':
        if not check_if_box_free_to_move(boxX+up,boxY-1,map,up):
            return False
        
    if map[boxX+up][boxY+1] == '[':
        if not check_if_box_free_to_move(boxX+up,boxY+1,map,up):
            return False
        
    return True

def move_box(boxX,boxY,map,up=-1):
    if map[boxX+up][boxY] == '.' and map[boxX+up][boxY+1] == '.': 
        map[boxX][boxY]      = '.'
        map[boxX][boxY+1]    = '.'
        map[boxX+up][boxY]   = '['
        map[boxX+up][boxY+1] = ']'
        return map 

    if map[boxX+up][boxY] == '[':
        map = move_box(boxX+up,boxY,map,up)
    
    if map[boxX+up][boxY] == ']':
        map = move_box(boxX+up,boxY-1,map,up)
        
    if map[boxX+up][boxY+1] == '[':
        map = move_box(boxX+up,boxY+1,map,up)

    return move_box(boxX,boxY,map,up)

while instructions:
    i = instructions.pop(0)
    match i:
        case '^':
            if not map[curr_r-1][curr_c] == '#':
                match map[curr_r-1][curr_c]:
                    case '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r-1][curr_c] = '@'
                        curr_r -= 1

                    case '[':
                        if check_if_box_free_to_move(curr_r-1,curr_c,map):
                            map = move_box(curr_r-1,curr_c,map)
                            map[curr_r][curr_c] = '.'
                            map[curr_r-1][curr_c] = '@'
                            curr_r -= 1

                    case ']':
                        if check_if_box_free_to_move(curr_r-1,curr_c-1,map):
                            map = move_box(curr_r-1,curr_c-1,map)
                            map[curr_r][curr_c] = '.'
                            map[curr_r-1][curr_c] = '@'
                            curr_r -= 1                   

        case '>':
            if not map[curr_r][curr_c+1] == '#':
                match map[curr_r][curr_c+1]:
                    case '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r][curr_c+1] = '@'
                        curr_c += 1
                    
                    case '[':
                        dc = 1
                        while map[curr_r][curr_c+dc] == '[': dc += 2
                        if map[curr_r][curr_c+dc] == '.':
                            map[curr_r][curr_c] = '.'
                            map[curr_r][curr_c+1] = '@'
                            for j in range(dc//2):
                                map[curr_r][curr_c+2+2*j] = '['
                                map[curr_r][curr_c+3+2*j] = ']'
                            curr_c += 1

        case 'v':
            if not map[curr_r+1][curr_c] == '#':
                match map[curr_r+1][curr_c]:
                    case '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r+1][curr_c] = '@'
                        curr_r += 1

                    case '[':
                        if check_if_box_free_to_move(curr_r+1,curr_c,map,1):
                            map = move_box(curr_r+1,curr_c,map,1)
                            map[curr_r][curr_c] = '.'
                            map[curr_r+1][curr_c] = '@'
                            curr_r += 1

                    case ']':
                        if check_if_box_free_to_move(curr_r+1,curr_c-1,map,1):
                            map = move_box(curr_r+1,curr_c-1,map,1)
                            map[curr_r][curr_c] = '.'
                            map[curr_r+1][curr_c] = '@'
                            curr_r += 1         

        case '<':
            if not map[curr_r][curr_c-1] == '#':
                match map[curr_r][curr_c-1]:
                    case '.':
                        map[curr_r][curr_c] = '.'
                        map[curr_r][curr_c-1] = '@'
                        curr_c -= 1
                    
                    case ']':
                        dc = 1
                        while map[curr_r][curr_c-dc] == ']': dc += 2
                        if map[curr_r][curr_c-dc] == '.':
                            map[curr_r][curr_c] = '.'
                            map[curr_r][curr_c-1] = '@'
                            for j in range(dc//2):
                                map[curr_r][curr_c-2-2*j] = ']'
                                map[curr_r][curr_c-3-2*j] = '['
                            curr_c -= 1
 
GPS = 0
for r, submap in enumerate(map):
    for c, element in enumerate(submap):
        if element == '[':
            GPS += 100*r + c
print('GPS coordinate: ', GPS)
```
